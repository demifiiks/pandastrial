## Cleaning


```python
travels_newdata = travels_data.copy() ## to create a new copy of the datasaet
```

```python
# to remove white spaces, we create a function

def remove_white_space(text):
    if type(text) == str:
        return text.strip()
    return text
```

```python
#removing the white space in the travel start date colulumn

travels_newdata['Travel Start Date'] = travels_newdata['Travel Start Date'].apply(remove_white_space)
```

```python
travels_newdata.info()
```

```python
travels_data.info()
```

```python
# changing the "travel start date" data type to date time

travels_newdata['Travel Start Date'] = pd.to_datetime(travels_newdata['Travel Start Date'])
```

```python
# changing the "travel end date" data type to date time

travels_newdata['Travel End Date'] = pd.to_datetime(travels_newdata['Travel End Date'])
```

```python
travels_newdata.info()
```

```python
## finding if null/nan values exists
travels_newdata[travels_newdata['Purpose Of Travel'].isna()]
```

```python
#filling null values in 'purpose of travel' column with 'not provided'
travels_newdata['Purpose Of Travel'] = travels_newdata['Purpose Of Travel'].fillna('Not Provided')
```

```python
travels_newdata.info()
```

```python
## creating functions to split a text
def extract_state(text):
    """EXTRACT STATE"""
    result = text.split(',')[1]
    return result

def extract_city(text):
    """EXTRACT CITY"""
    result = text.split(',')[0]
    return result
```

```python
## to split the destination column into state and city
travels_newdata['State'] = travels_newdata['Destination'].apply(extract_state)
travels_newdata['City'] = travels_newdata['Destination'].apply(extract_city)
```

```python
travels_newdata
```

```python
def extract_Firstname(text):
    """EXTRACT FIRST NAME"""
    result = text.split(' ')[0]
    return result


def extract_lastname_or_initials(text):
    """Extracts Last Name or Middle Initials and Last Name"""
    result = text.split(' ')
    if len(result) > 2:
        join_name = ' '.join(result[1:])
        return join_name
    else:
        return result[-1]
    

    

    
    
```

```python
travels_newdata['First Name'] = travels_newdata['Employee'].apply(extract_Firstname)
travels_newdata['Last Name'] = travels_newdata['Employee'].apply(extract_lastname_or_initials)
```

```python
# function to remove special characters
def remove_special_characters(name):
    result = name.replace('@', ' ')
    result = result.replace('!', ' ')

    return result
```

```python
# removing the special characters

travels_newdata['Employee'] = travels_newdata['Employee'].apply(remove_special_characters)
travels_newdata['First Name'] = travels_newdata['First Name'].apply(remove_special_characters)
travels_newdata['Last Name'] = travels_newdata['Last Name'].apply(remove_special_characters)
```
